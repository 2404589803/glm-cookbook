# Use GLM-4 Models in Microsoft TaskWeaver Framework

TaskWeaver 是一个代码优先的 Agent 框架，用于无缝规划和执行数据分析任务。这种创新框架通过代码片段解释用户请求，并以函数的形式有效协调各种插件，以有状态的方式执行数据分析任务。
本代码将使用 微软的 TaskWeaver 框架，并使用 GLM-4 模型来完成具有规划性的 Agent 任务。

## 安装 TaskWeaver

请注意，本 cookbook 需要在 TaskWeaver 仓库内运行，基础上进行修改，因此，我们需要先安装 TaskWeaver，你可以按照[官方方案](https://microsoft.github.io/TaskWeaver/docs/experience) 进行安装。

```shell
git clone https://github.com/microsoft/TaskWeaver.git
cd TaskWeaver
# install the requirements
pip install -r requirements.txt
```

## 配置 GLM-4 模型


接着，我们只需要修改 `project/taskweaver_config.json` 中的相关配置，按照[官方文档](https://microsoft.github.io/TaskWeaver/docs/llms/glm) 进行配置，即可使用 GLM-4 模型作为模型底座。

```json
{
  "llm.api_base": "https://open.bigmodel.cn/api/paas/v4",
  "llm.api_key": "your keys",
  "llm.api_type": "zhipuai",
  "llm.model": "glm-4",
  "llm.embedding_model": "embedding-2",
  "llm.embedding_api_type": "zhipuai"
}
```

我们使用命令来启动 TaskWeaver 服务，同时，我们执行[官方示例](https://microsoft.github.io/TaskWeaver/docs/experience) 来验证 GLM-4 的规划和执行能力。
