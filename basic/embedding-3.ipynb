# Generating Text Embeddings with Embedding-3

**This tutorial is available in English and is attached below the Chinese explanation**

Embedding-3 是文本向量模型，将输入的文本信息进行向量化表示，以便于结合向量数据库为大模型提供外部知识库，模型支持自定义向量维度，建议选择256、512、1024或2048维度。

## 1. Set up the API Key

首先，设置环境变量并初始化 ZhipuAI client。

First, set up an environmental variable and initialize the ZhipuAI client.

```python
from zhipuai import ZhipuAI

client = ZhipuAI(api_key="your api key")
```

## 2. Make the API Request

接下来，我们为一系列文本输入生成嵌入模型。模型参数应设置为 “embedding-3”。输入是一个文本字符串列表，您希望将其转换为嵌入式编码。列表中的每个字符串都代表一个将由模型处理的独立输入。

Next, you will generate embeddings for a list of text inputs. The model parameter should be set to "embedding-3". The input is a list of text strings that you want to convert into embeddings. Each string in this list represents a separate input that will be processed by the model.

```python
response = client.embeddings.create(
    model="embedding-3",
     input=[
        "美食非常美味，服务员也很友好。",
        "这部电影既刺激又令人兴奋。",
        "阅读书籍是扩展知识的好方法。"
    ],
)
```

## 3. Print the Embeddings

最后，打印嵌入。每个索引（0、1、2）都对应于输入列表中由原始自然语言编码的嵌入列表。

Lastly, print the embeddings. Each index (0, 1, 2) corresponds to the list of embedding that was encoded from the original natural language in the input list.

```python
print(response)
```
